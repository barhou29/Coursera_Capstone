```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
path = requests.get("https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M")
sp = BeautifulSoup(path.content, 'html.parser')
```

```python
tab = sp.find('tbody')
rows = tab.select('tr')
row = [r.get_text() for r in rows]
```

```python
df = pd.DataFrame(row)
df1 = df[0].str.split('\n', expand=True)
df2 = df1.rename(columns=df1.iloc[0])
df3 = df2.drop(df2.index[0])
df3.head()
```

```python
df4 = df3[df3.Borough != 'Not assigned']
df4.head()
```

```python
df_final = df4.replace("Not assigned", "Queen's Park")
df_final.head()
```
